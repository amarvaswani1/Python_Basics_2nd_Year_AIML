```python
# Creating dictionary
student_data = {
    "Student1": ["Aarav Sharma", 20, "Artificial Intelligence", "Computer Science"],
    "Student2": ["Priya Verma", 19, "Data Science", "Information Technology"],
    "Student3": ["Rahul Mehta", 21, "Cyber Security", "Computer Science"],
    "Student4": ["Sneha Patel", 20, "Cloud Computing", "Information Technology"],
    "Student5": ["Vikram Singh", 22, "Machine Learning", "Artificial Intelligence"],
    "Student6": ["Ananya Gupta", 19, "Full Stack Development", "Computer Science"],
    "Student7": ["Rohan Das", 21, "DevOps", "Information Technology"],
    "Student8": ["Neha Joshi", 20, "Software Development", "Computer Science"]
}
```

```python
# Printing all students data
print(student_data)
```

```python
# Printing seperate student data
print(student_data["Student1"])

# Printing only specific data like name/age
print(student_data["Student1"][0])
print(student_data["Student1"][1])
```

```python
# Using for loop to print details of every student
for student, details in student_data.items():
    print("Student ID:", student)
    print("Name:", details[0])
    print("Age:", details[1])
    print("Domain:", details[2])
    print("Department:", details[3])
    print()
```

```python
# Single dictionary with keys having multiple values
student_data = {
    "name": [
        "Aarav Sharma", "Priya Verma", "Rahul Mehta", "Sneha Patel",
        "Vikram Singh", "Ananya Gupta", "Rohan Das", "Neha Joshi"
    ],
    "age": [20, 19, 21, 20, 22, 19, 21, 20],
    "domain": [
        "Artificial Intelligence", "Data Science", "Cyber Security", "Cloud Computing",
        "Machine Learning", "Full Stack Development", "DevOps", "Software Development"
    ],
    "department": [
        "Computer Science", "Information Technology", "Computer Science", "Information Technology",
        "Artificial Intelligence", "Computer Science", "Information Technology", "Computer Science"
    ]
}
```

```python
# Printing all students data
print(student_data)
```

```python
# For accessing names in the above dictionary
print(student_data['name'])

# For accessing age in the above dictionary 
print(student_data["age"])
```

```python
# For accessing specific data 
# First student name
print(student_data["name"][0])

# Second student age 
print(student_data["age"][1])

# Third student domain
print(student_data["domain"][1])

# Fourth student department
print(student_data["department"][3])
```

```python
# Using for loop to print details of every student
for i in range(len(student_data["name"])):
    print(student_data["name"][i])
    print(student_data["age"][i])
    print(student_data["domain"][i])
    print(student_data["department"][i])
    print()
```

```python
# Single dictionary consisting multiple nested dictionary
student_data = {
    "Student1": {
        "name": "Aarav Sharma",
        "age": 20,
        "domain": "Artificial Intelligence",
        "department": "Computer Science"
    },
    "Student2": {
        "name": "Priya Verma",
        "age": 19,
        "domain": "Data Science",
        "department": "Information Technology"
    },
    "Student3": {
        "name": "Rahul Mehta",
        "age": 21,
        "domain": "Cyber Security",
        "department": "Computer Science"
    },
    "Student4": {
        "name": "Sneha Patel",
        "age": 20,
        "domain": "Cloud Computing",
        "department": "Information Technology"
    },
    "Student5": {
        "name": "Vikram Singh",
        "age": 22,
        "domain": "Machine Learning",
        "department": "Artificial Intelligence"
    },
    "Student6": {
        "name": "Ananya Gupta",
        "age": 19,
        "domain": "Full Stack Development",
        "department": "Computer Science"
    },
    "Student7": {
        "name": "Rohan Das",
        "age": 21,
        "domain": "DevOps",
        "department": "Information Technology"
    },
    "Student8": {
        "name": "Neha Joshi",
        "age": 20,
        "domain": "Software Development",
        "department": "Computer Science"
    }
}
```

```python
# Printing all students data
print(student_data)
```

```python
# Printing specific data 
# First student name
print(student_data["Student1"]["name"])

# Second student age 
print(student_data["Student2"]["age"])

# Fifth student domain
print(student_data["Student5"]["domain"])

# Seventh student department
print(student_data["Student7"]["department"])

# Printing specific student details
print(student_data["Student2"])

# Printing names 
for student in student_data:
    print(student_data[student]["name"])
```

```python
# Using for loop to print details
# For printing details of every student
for student, details in student_data.items():
    print("Student:", student)
    print("Name:", details["name"])
    print("Age:", details["age"])
    print("Domain:", details["domain"])
    print("Department:", details["department"])
    print()

# For printing all names
for student in student_data:
    print(student_data[student]["name"])

# For printing students in computer science department
for student, details in student_data.items():
    if details["department"] == "Computer Science":
        print(details["name"])
```
